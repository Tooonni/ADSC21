## Transfersumme und Anzahl der Transfers pro Spieler

```python
players["number_of_transfers"] = 0
players["total_transfer_fee"] = 0

transfers["transfer_fee"] = transfers["transfer_fee"].fillna(0.0)

for player in players["player_id"].to_list():
    number_of_transfers = 0
    total_transfer_fee = 0

    for transfer in transfers[transfers["player_id"] == player].itertuples():
        number_of_transfers += 1
        total_transfer_fee += transfer.transfer_fee

    players.loc[players["player_id"] == player, "number_of_transfers"] = number_of_transfers
    players.loc[players["player_id"] == player, "total_transfer_fee"] = total_transfer_fee
```

## Summe von Startelf, Einwechslung, Kapitän & Spiele

```python
game_lineups["type"].value_counts().to_list
```

```python
game_lineups["team_captain"].value_counts().to_list
```

```python
players["total_starting_lineups"] = 0
players["total_substitute_appearances"] = 0
players["total_captain_appearances"] = 0
players["total_games"] = 0

for player in players["player_id"].to_list():
    total_starting_lineups = 0
    total_substitute_appearances = 0
    total_captain_appearances = 0
    total_games = 0

    for game_lineup in game_lineups[game_lineups["player_id"] == player].itertuples():
        if game_lineup.type == "starting_lineup":
            total_starting_lineups += 1
        elif game_lineup.type == "substitutes":
            total_substitute_appearances += 1
        else:
            None

        if game_lineup.team_captain == 1:
            total_captain_appearances += 1
        else:
            None
    
        total_games += 1

        players.loc[players["player_id"] == player, "total_starting_lineups"] = total_starting_lineups
        players.loc[players["player_id"] == player, "total_substitute_appearances"] = total_substitute_appearances
        players.loc[players["player_id"] == player, "total_captain_appearances"] = total_captain_appearances
        players.loc[players["player_id"] == player, "total_games"] = total_games
```

## Summe von Gelben Karten, Rote Karten, Tore, Vorlagen & gespiele Minuten

```python
players["total_yellow_cards"] = 0
players["total_red_cards"] = 0
players["total_goals"] = 0
players["total_assists"] = 0
players["total_minutes_played"] = 0

for player in players["player_id"].to_list():
    total_yellow_cards = 0
    total_red_cards = 0
    total_goals = 0
    total_assists = 0
    total_minutes_played = 0

    for appearance in appearances[appearances["player_id"] == player].itertuples():
        total_yellow_cards += appearance.yellow_cards
        total_red_cards += appearance.red_cards
        total_goals += appearance.goals
        total_assists += appearance.assists
        total_minutes_played += appearance.minutes_played

        players.loc[players["player_id"] == player, "total_yellow_cards"] = total_yellow_cards
        players.loc[players["player_id"] == player, "total_red_cards"] = total_red_cards
        players.loc[players["player_id"] == player, "total_goals"] = total_goals
        players.loc[players["player_id"] == player, "total_assists"] = total_assists
        players.loc[players["player_id"] == player, "total_minutes_played"] = total_minutes_played
```
